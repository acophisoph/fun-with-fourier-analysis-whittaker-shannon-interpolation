```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

# Number of samplepoints
N = 600
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
y = np.sin(50.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf = scipy.fftpack.fft(y)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)

fig, ax = plt.subplots()
plt.title('FFT transform')
ax.set_xlabel('Frequency')
ax.set_ylabel('Energy')
ax.plot(xf, 2.0/N * np.abs(yf[:N//2]))

fig, ax = plt.subplots()
plt.title('Function graph')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x, y, 'g')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import sympy as sp
from sympy import symbols, Eq

# Points to create smooth graph
n = 600
t = 1

# variables to change for the source function
a = 1.0
b = 20.0
c = 1.0
d = 10.0



# Function to interpolate on
x = np.linspace(-t, t, n)
y = a * np.sin(b * np.pi*x) + c * np.sin(d * np.pi*x)

# Number of sample points
ni = 100

# Points to interpolate on
xp = np.linspace(-t, t, ni)
yp = a * np.sin(b * np.pi*xp) + c * np.sin(d * np.pi*xp)

# Whittaker-Shannon Interpolation Initialize
w = symbols('w') # initialize variable for interpolation formula x(w) = WS_Interpolation formula
yi = 0 #intialize equation, yi = WS_Interpolation formula
s = 0 #count what the current step is

# Actual Interpolation Function
for item in xp:
    yi = yi + item * sp.sinc((w - (s*(2 * t/ni)))/(2 * t/ni))
    s += 1
f = sp.lambdify(w, yi, "numpy") #change expression to one that numpy can evaluate
yf = f(x)
xf = x # redundant code, just made to match the variables



fig, ax = plt.subplots()
plt.title('Function graph')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x, y, 'r')

fig, ax = plt.subplots()
plt.title('Whittaker Shannon Interpolation points')
ax.set_xlabel('xp')
ax.set_ylabel('yp')
plt.scatter(xp, yp)

fig, ax = plt.subplots()
plt.title('Reconstructed Graph')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(xf, yf, 'b')

plt.show()
```
